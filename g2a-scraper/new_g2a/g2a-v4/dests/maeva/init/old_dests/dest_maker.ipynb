```python
from datetime import datetime, timedelta
import json
import pandas as pd
```

```python
start_date = "28/03/2024"
end_date = "26/10/2024"
dest_url_path = "./dest12.json"
output_path = "./../new_dests/new_dest_25_03_2024/"
name = "old_dest12.json"
```

```python
start_date = datetime.strptime(start_date, "%d/%m/%Y").strftime('%Y-%m-%d')
end_date = datetime.strptime(end_date, "%d/%m/%Y").strftime('%Y-%m-%d')
```

```python
urls = []
with open(dest_url_path, 'r') as openfile:
    urls = json.load(openfile)
len(urls)
```

```python
formated_urls = []
```

```python
dates = pd.bdate_range(start=start_date, end=end_date, freq='C', weekmask='Sat').strftime('%Y-%m-%d').to_list()
```

```python
dates
```

```python
for date in dates:
    next_saturday = datetime.strptime(date, '%Y-%m-%d')
    date_debut = datetime.strftime(next_saturday, '%Y-%m-%d')
    date_fin = datetime.strftime(
        next_saturday + timedelta(days=6), '%Y-%m-%d')
    for url in urls:
        dest_id = url.split('_')[1].split('.')[0]
        formated_urls.append(f"{url}?date_debut={date_debut}&date_fin={date_fin}&residence_cle={dest_id}&formule=0&ordreSeo=prixAsc")
        next_saturday += timedelta(days=7)
```

```python
formated_urls
```

```python
len(formated_urls)
```

```python
with open(f"{output_path}{name}", 'w') as openfile:
    openfile.write(json.dumps(formated_urls, indent=4))
```
